```python
def Horner_Method(coefficients, x):
    result = 0
    for coefficient in reversed(coefficients):
        result = result * x + coefficient
    return result

def Bisection_Method(coefficients, a, b, epsilon, max_iter=1000, iteration=0):
    if iteration > max_iter:
        raise RecursionError("Maximum iterations reached. The method may not be converging.")

    p = (a + b) / 2
    p1 = Horner_Method(coefficients, p)

    # Check if p1 is close enough to zero
    if abs(p1) <= epsilon:
        return p

    # Check if the sign changes between a and p or p and b
    f_a = Horner_Method(coefficients, a)

    if f_a * p1 < 0:
        return Bisection_Method(coefficients, a, p, epsilon, max_iter, iteration + 1)
    else:
        return Bisection_Method(coefficients, p, b, epsilon, max_iter, iteration + 1)

# Example usage
coefficients = [-25,0,0,1]  
a = 2
b = 3
epsilon = 0.001

try:
    result = Bisection_Method(coefficients, a, b, epsilon)
    print(f"The root of the polynomial is approximately {result}")
except RecursionError as e:
    print(e)
```

```python
def newtons_method(f, f_prime, p0, n):
    p = p0
    for _ in range(n):
        if p - (p - f(p) / f_prime(p)) < 0.0001:
            return p
        p = p - f(p) / f_prime(p)
    return p

# Example usage:
if __name__ == "__main__":
    # Example: Find the square root of 2 by solving f(x) = x^2 - 2 = 0
    f = lambda x: x**3 - 2*x**2 - 5
    f_prime = lambda x: 3*x**2 - 4*x
    initial_guess = 3
    iterations = 10000
    zero_approx = newtons_method(f, f_prime, initial_guess, iterations)
    f_val = f(zero_approx)
    print("Approximate zero:", zero_approx)
    print("Value at x = ", zero_approx, "is f(x) = " , f_val)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fixed_point_iteration(g, p0, n, print_iterations=False):
    """
    Perform fixed point iteration for function g starting at p0 for n iterations.
    Optionally prints each iterate.
    Returns the list of iterates.
    """
    iterates = [p0]
    current = p0
    for i in range(n):
        current = g(current)
        iterates.append(current)
        if print_iterations:
            print(f"Iteration {i+1}: {current}")
    return iterates

def g_factory(mu):
    """
    Returns the function g(x) = mu*(0.5 - |x - 0.5|) for a given mu.
    """
    def g(x):
        return mu * (0.5 - np.abs(x - 0.5))
    return g

# Starting value: x0 = /6
p0 = np.pi/6

# List of  values to test
mus = [1.1, 1.5]

# Loop over the two  values
for mu in mus:
    print(f"\n=== Fixed point iteration for  = {mu} with 10 iterations ===")
    g = g_factory(mu)
    
    # 10 iterations: print intermediate values
    iterates_10 = fixed_point_iteration(g, p0, 10, print_iterations=True)
    
    # Now do 10^4 iterations and store the iterates (without printing)
    iterates_long = fixed_point_iteration(g, p0, 10**4)
    
    # Plot the iterates vs. iteration number
    plt.figure(figsize=(10, 4))
    plt.plot(iterates_long, lw=0.5)
    plt.xlabel("Iteration")
    plt.ylabel("x")
    plt.title(f"Fixed Point Iteration ( = {mu})")
    plt.grid(True)
    plt.show()
```

```python
import math
import numpy as np

def f(x):
    return np.polyval([600, -550, 200, -20, -1], x) 

def df(x):
    # Derivative: f'(x) =  * sec^2(x)
    return np.polyval([2400, -1650, -400, -20], x)

# ---------------------------
# Bisection Method
# ---------------------------
def bisection(f, a, b, n):
    iterates = []
    for i in range(n):
        c = (a + b) / 2.0
        iterates.append(c)
        if f(c) == 0:
            return c, iterates
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c, iterates

# ---------------------------
# Newton's Method
# ---------------------------
def newton(f, df, p, n):
    iterates = [p]
    for i in range(n):
        p = p - f(p) / df(p)
        iterates.append(p)
    return p, iterates

# ---------------------------
# Secant Method
# ---------------------------
def secant(f, p0, p1, n):
    iterates = [p0, p1]
    for i in range(2, n + 2):  # Already have two values.
        denominator = f(p1) - f(p0)
        if denominator == 0:
            break  # Avoid division by zero.
        p = p1 - f(p1) * (p1 - p0) / denominator
        iterates.append(p)
        p0, p1 = p1, p
    return p, iterates

# ---------------------------
# Main Comparison
# ---------------------------
if __name__ == "__main__":
    N = 10  # Number of iterations

    # Bisection: Use interval [0, 0.48] (f(0) < 0, f(0.48) > 0)
    root_bis, iters_bis = bisection(f, 0, 0.48, N)

    # Newton's Method: initial guess p0 = 0 (given), though it will perform badly.
    root_newt, iters_newt = newton(f, df, 0, N)

    # Secant Method: starting values p0 = 0, p1 = 0.48 (as given)
    root_sec, iters_sec = secant(f, 0, 0.48, N)

    # Print results
    print("Bisection Method Iterates:")
    for i, x in enumerate(iters_bis):
        print(f"  Iteration {i+1}: {x:.9f}")
    print("Approximate root (Bisection):", root_bis)
    print()

    print("Newton's Method Iterates (starting at 0):")
    for i, x in enumerate(iters_newt):
        print(f"  Iteration {i+1}: {x:.9f}")
    print("Approximate root (Newton):", root_newt)
    print()

    print("Secant Method Iterates:")
    for i, x in enumerate(iters_sec):
        print(f"  Iteration {i+1}: {x:.9f}")
    print("Approximate root (Secant):", root_sec)
    print()

    true_root = (1 / math.pi) * math.atan(6)
    print("True root (approx.):", true_root)
```
